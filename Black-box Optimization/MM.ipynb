```python
import pandas as pd
import numpy as np
from tslearn.clustering import TimeSeriesKMeans
import matplotlib.pyplot as plt

# 1. 데이터 로드
train_file_path = 'data/train.csv'
test_file_path = 'data/test.csv'

train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

# 2. 데이터 전처리
# ID 열 제거하고 나머지 데이터를 시계열 데이터로 사용
X_train = train_data.drop(columns=['ID', 'y']).values
X_test = test_data.drop(columns=['ID']).values

# 시계열 데이터로 변환하기 위해 차원을 확장
X_train = np.expand_dims(X_train, axis=1)
X_test = np.expand_dims(X_test, axis=1)

# 3. 모델 설정 및 학습
# TimeSeriesKMeans 모델 설정 (클러스터 개수는 데이터에 맞게 조정 가능)
n_clusters = 3  # 임의의 클러스터 개수 설정
model = TimeSeriesKMeans(n_clusters=n_clusters, metric="dtw", max_iter=10, verbose=True)
```

```python
# 모델 학습 (훈련 데이터로 학습)
model.fit(X_train)
```

```python

# 4. 훈련 데이터와 테스트 데이터에 대한 클러스터 예측
train_clusters = model.predict(X_train)
test_clusters = model.predict(X_test)

# 5. 결과 확인
print("Train Clusters (Top 10):", train_clusters[:10])
print("Test Clusters (Top 10):", test_clusters[:10])

# 6. 결과 시각화 (클러스터링된 시계열 시각화)
for cluster_id in range(n_clusters):
    plt.figure()
    for i in range(len(X_train)):
        if train_clusters[i] == cluster_id:
            plt.plot(X_train[i].ravel(), "k-", alpha=0.2)
    plt.title(f"Cluster {cluster_id}")
    plt.show()
```
