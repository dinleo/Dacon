```python
import pandas as pd
import h2o
from h2o.automl import H2OAutoML


# 파일 경로
train_path = 'data/train.csv'
test_path = 'data/test.csv'
```

```python
# H2O 클러스터 시작
h2o.init()
```

```python
# 훈련 및 테스트 데이터 로드 (H2OFrame으로 변환)
train = h2o.import_file(train_path)
test = h2o.import_file(test_path)

# 타깃 변수 지정 (예: 'target'이 타깃 변수일 경우)
y = 'y'  # 타깃 변수 이름을 지정하세요
x = train.columns  # 모든 열을 독립 변수로 설정
x.remove(y)  # 타깃 변수는 독립 변수에서 제외
x.remove('ID')
```

```python
# AutoML 모델 생성
aml = H2OAutoML(max_runtime_secs=3000, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

```python
# 테스트 데이터 예측
preds = aml.leader.predict(test)
```

```python
print(aml.leader)

# 모델 랭킹 보기
lb = aml.leaderboard
lb.head()  # 상위 10개 모델 출력
```

```python
# sample_submission.csv 파일 불러오기
submission_file_path = 'data/sample_submission.csv'
submission_data = pd.read_csv(submission_file_path)

# 예측 결과를 sample_submission 형식에 맞춰 'y' 컬럼에 삽입
submission_data['y'] = preds

# 결과를 CSV로 저장 (경로는 '/mnt/data/submission_result.csv')
output_path = 'sub.csv'
submission_data.to_csv(output_path, index=False)

output_path  # 출력 경로 확인
```
